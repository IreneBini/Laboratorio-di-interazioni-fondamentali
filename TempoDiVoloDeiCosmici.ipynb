## MISURA DEL TEMPO DI VOLO DEI RAGGI COSMICI

L'obiettivo di questa esperienza è la misura del tempo di volo dei raggi cosmici che raggiungono la superficie terrestre (in particolare muoni). Per farlo, si determina la distribuzione di essi al livello del mare e di conseguenza la distribuzione in velocità.Metteremo in evidenza la piccola frazione di essi che raggiungono velocità relativistiche.

## APPARATO SPERIMENTALE
1. ODOMETRO, BICRON BC408: Barra di scintillatore plastico a sezione quadrata (4 x 4 cm) e di lunghezza circa 2 m. Viene letta alle due estremità da due PMT (Photomultiplier Tube: PMT1 e PMT2).
2. Fotomoltiplicatore mobile (PMT3) con scintillatore plastico di piccole dimensioni. Si può posizionare sopra la barra in vari punti.
4. Scheda di acquisizione *Domino Ring Sampler* (DRS4) che consente di digitalizzare fino a 4 segnali analogici con frquenza massima 5 Gs/s. **Limiti di tensione**: segnali che può acquisire devono avere ampiezza massima di 1V picco-picco.
5. Moduli logici NIM (ritardi, contatori, .. )
6. Oscilloscopio per la calibrazione ed il debug dell’apparato.



## Strategia
La differenza temporale dei segnali dei due fotomoltiplicatori alle estremità della barra ci permetterà di determinare la posizione in cui il raggio cosmico ha attraversato la barra. Un terzo scintillatore
posto in basso ci permetterà infine di determinare un secondo punto della
traiettoria e da qui di ricostruire la direzione di provenienza del raggio cosmico.
Dovremo prestare molta attenzione alla differente efficienza di rivelazione al variare
dell’angolo di incidenza.

Selezioniamo gli eventi in cui il raggio cosmico attraversi sia la paletta scintillante
inferiore che la barra scintillante superiore e produca segnali da ognuno dei tre
fotomoltiplicatori. Dalla differenza dei tempi dei segnali prodotti dai PMT#1 e
PMT#2 ricostruiremo la posizione in cui il raggio cosmico ha attraversato la barra
superiore e da questo ricostruiremo l’angolo di incidenza .
Dalla misura del tempo di volo del raggio cosmico potremo anche stimare la
velocità della particella.

## Utilità

La differenza di tempo tra i segnali registrati da PMT1 e PMT2 fornisce una misura della posizione in cui la barra è stata attraversata da una particella ionizzante (a meno di una costante).
A causa dell'attenuazione della luce nella barra, anche il rapporto tra altezze di impulso fornisce informazioni sulla posizione in cui la barra è stata colpita.

PM3 si utilizza per misure di tempo di volo e per calibrazione (per esempio si può usare per rimuovere i termini costanti ignoti nelle misure di differenze di tempo).

```python
# soglia del secondo discriminatore = -39.7 mV
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import csv
from scipy.optimize import curve_fit

# Inserisco i dati
Alimentazione2 = np.array([1.2, 1.251, 1.299, 1.346, 1.348, 1.402, 1.449, 1.501, 1.553, 1.598, 1.649, 1.702, 1.746, 1.798]) # kV

# Questi li lascio invariati, vado a variare solo il secondo
V1 = 1.498 #kV
V3 = 1.502  #kV

# Durata dell'acquisizione
tt = 100 # s

# nano secondi
ns = 1e-9

# utili per il calcolo delle accidentali
omega_1 = 100 # ns larghezza del primo segnale discriminato
omega_2 = 100 # ns  larghezza del secondo segnale
omega_3 = 50 # ns  larghezza del terzo segnale
delta_t = 2 # ns scelto arbitrariamente

# conteggi
cont_1 = np.array([13052, 13008, 12713, 13036, 12767, 12956, 13065, 12850, 12911, 13215, 13036, 13019, 12905, 12920])
cont_2 = np.array([2087, 2941, 3951, 5204, 5419, 7538, 9694, 12780, 16722, 20532, 25413, 31214, 37179, 44793])
cont_3 = np.array([997, 990, 987, 1055, 953, 989, 1000, 1021, 1020, 991, 996, 969, 1053, 990])
cont_13 = np.array([47, 46, 54, 61, 33, 47, 61, 54, 47, 54, 55, 45, 57, 47])
cont_123 = np.array([36, 40, 47, 50, 32, 42, 59, 53, 47, 54, 55, 44, 57, 47])
cont_12 = np.array([1911, 2300, 2659, 3070, 3122, 3780, 4382, 5300, 6508, 7807, 8693, 9787, 10338, 10957])

# Calcolo errore poissoniano su conteggi
cont_1_err = np.ones(len(cont_2))
cont_2_err = np.ones(len(cont_2))
cont_3_err = np.ones(len(cont_2))
cont_13_err = np.ones(len(cont_2))
cont_123_err = np.ones(len(cont_2))
for i in range(0, len(cont_2)):
    cont_1_err[i] = np.sqrt(cont_1[i]) # è un array
    cont_2_err[i] = np.sqrt(cont_2[i])
    cont_3_err[i] = np.sqrt(cont_3[i])
    cont_13_err[i] = np.sqrt(cont_13[i])
    cont_123_err[i] = np.sqrt(cont_123[i])

# calcolo le coincidenze accidentali e l'efficienza
acc = np.ones(len(cont_1))
e2 = np.ones(len(cont_1))
err_e2 = np.ones(len(cont_1))
for i in range(0, len(cont_1)):
    acc[i] = (cont_1[i]/tt)*(cont_3[i]/tt)*(omega_1+omega_3-2*delta_t)*ns
    e2[i] = ((cont_123[i]/tt)-acc[i])/(cont_13[i]/tt)
    err_e2[i] = np.sqrt(e2[i]*(1.- e2[i])/(cont_13[i]))
    # divido per il numero di conteggi ovvero rate per durata acquisizione

print('Stampo i dati inseriti nel grafico')
for i in range(0, len(cont_1)):
    print('iterazione %d' % i)
    print('coincidenze accidentali: %.5f' % acc[i])
    print('l efficienza alla tensione di alimentazione %.3f kV vale: %.3f +- %.3f' % (Alimentazione2[i], e2[i], err_e2[i]))


## grafico efficienza
x2=Alimentazione2
x1=cont_2/tt
y=e2

fig, ax1 = plt.subplots()

plt.title('$\epsilon 2$ vs $V_2$ ($V_{thr}=-39.7$ mV e $x=140$ cm)', fontsize = 24)
matplotlib.rc('xtick', labelsize=18)
matplotlib.rc('ytick', labelsize=18)
plt.grid(True)


ax1.set_xlabel(r"$V_2$ [kV]", fontsize = 18)
ax1.set_ylabel('$\epsilon_2$', fontsize = 18)
ax1.errorbar(x2, y, yerr=err_e2, marker='.', color='cornflowerblue', linestyle = 'none', label = 'data')

for i in range(0, len(x1)): # i è l'indice di x1
    txt = x1[i] # txt è il contenuto di x1
    ax1.annotate(txt, (x2[i], y[i]), color='k', size=16)

plt.ylim(0.7, 1.01)

plt.legend(loc = 'upper left', fontsize = 16)

plt.show()
```
